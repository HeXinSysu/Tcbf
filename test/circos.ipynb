```python
!mkdir pycircos
%cd pycircos
!mkdir sample_data
%cd sample_data
#The following example data was downloaded from https://venyao.xyz/shinyCircos/.
import requests
def wget_data(url):
    local_filename = url.split("/")[-1]
    response = requests.get(url)
    with open(local_filename,"wb")as f:
        for chunk in response.iter_content(chunk_size=8192 ):
            f.write(chunk)
    return local_filename
```

```python

file1 = """https://github.com/ponnhide/pyCircos-examples/raw/main/example_notebooks/sample_data/example_data_barplot.csv
https://github.com/ponnhide/pyCircos-examples/raw/main/example_notebooks/sample_data/example_data_chromosome_cytoband.csv
https://github.com/ponnhide/pyCircos-examples/raw/main/example_notebooks/sample_data/example_data_chromosome_general.csv
https://github.com/ponnhide/pyCircos-examples/raw/main/example_notebooks/sample_data/example_data_links.csv
https://github.com/ponnhide/pyCircos-examples/raw/main/example_notebooks/sample_data/example_data_point.csv
https://github.com/ponnhide/pyCircos-examples/raw/main/example_notebooks/sample_data/example_data_rect_gradual.csv""".split()
for f in file1:
    wget_data(f)

```
